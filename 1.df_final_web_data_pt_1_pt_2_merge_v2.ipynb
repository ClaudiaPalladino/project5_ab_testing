```python
# Drop row with "variation" nulls
full_merged_exp = full_merged[full_merged['variation'].notna()]
print("Rows in experiment (Control/Test):", len(full_merged_exp))
```

```python
# Drop rows with missing demographics
full_merged_exp_demo = full_merged_exp.dropna(subset=['clnt_age','gender','num_accts','bal'])

full_merged_exp_demo.info()
```

```python
# Make a proper copy of full_merged_exp_demo
full_merged_exp_demo = full_merged_exp_demo.copy()
```

```python
# Convert IDs to strings to avoid concatenation issues
full_merged_exp_demo['client_id'] = full_merged_exp_demo['client_id'].astype(str)
full_merged_exp_demo['visit_id']  = full_merged_exp_demo['visit_id'].astype(str)

# Create a session-level primary key
full_merged_exp_demo['session_id'] = (
    full_merged_exp_demo['client_id'] + "_" + full_merged_exp_demo['visit_id']
)

# Check: number of unique sessions
print("Unique sessions:", full_merged_exp_demo['session_id'].nunique())
print("Rows:", len(full_merged_exp_demo))
```

```python
full_merged_exp_demo.head(2)
```

```python
import pandas as pd
import numpy as np

# gender var - Standardize, clean, and label
gender_map = {   # value transformation via dictionary
    'M': 'male',
    'F': 'female'
}

full_merged_exp_demo['gender_clean'] = (   # Creates a new column if it does not already exist (preserves raw data: best practice in data cleaning)
    full_merged_exp_demo['gender']
    .str.upper()                 # normalize case
    .replace(['U', 'X'], np.nan) # treat U and X as missing
    .map(gender_map)             # apply labels
)

# Convert to categorical
full_merged_exp_demo['gender_clean'] = pd.Categorical(
    full_merged_exp_demo['gender_clean'],
    categories=['male', 'female']
)

print(full_merged_exp_demo)
```

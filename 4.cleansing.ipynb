```python
df = df.groupby(['client_id', 'visitor_id', 'visit_id', 'process_step']).max().reset_index()
```

```python
# Pivot the dataframe to make each process_step a column with date_time as value
df_pivoted = df.pivot_table(
    index=['client_id', 'visitor_id', 'visit_id'],
    columns='process_step',
    values='date_time',
    aggfunc='first'
).reset_index()

# Flatten column names
df_pivoted.columns.name = None

df_pivoted
```

```python
# Pivot the dataframe to make each process_step a column with date_time as value
df_pivoted = df.pivot_table(
    index=['client_id', 'visitor_id', 'visit_id'],
    columns='process_step',
    values='date_time',
    aggfunc='first'
).reset_index()

# Flatten column names
df_pivoted.columns.name = None

df_pivoted
```

```python
df = df_pivoted[['client_id', 'visitor_id', 'visit_id', 'start', 'step_1', 'step_2', 'step_3', 'confirm']]
```

```python
df.to_csv('a_b_testing.csv', index=False)
```

```python
df = df[
    (df['start'].notna() & df['step_1'].notna() & df['step_2'].notna() & df['step_3'].notna() & df['confirm'].notna())
    | (df['start'].notna() & df['step_1'].notna() & df['step_2'].notna() & df['step_3'].notna())
    | (df['start'].notna() & df['step_1'].notna() & df['step_2'].notna())
    | (df['start'].notna() & df['step_1'].notna())
    | (df['start'].notna())
]
df.to_csv('a_b_testing_2.csv', index=False)
```

```python
# Convert datetime columns to datetime type
step_cols = ['start', 'step_1', 'step_2', 'step_3', 'confirm']
for col in step_cols:
    df[col] = pd.to_datetime(df[col])

# Calculate deltas in seconds
df['delta_start_to_step1'] = (df['step_1'] - df['start']).dt.total_seconds()
df['delta_step1_to_step2'] = (df['step_2'] - df['step_1']).dt.total_seconds()
df['delta_step2_to_step3'] = (df['step_3'] - df['step_2']).dt.total_seconds()
df['delta_step3_to_confirm'] = (df['confirm'] - df['step_3']).dt.total_seconds()

df
```

```python
df.columns
```

```python
df = df[[
    'client_id', 'visitor_id', 'visit_id',
    'delta_start_to_step1', 'delta_step1_to_step2',
    'delta_step2_to_step3', 'delta_step3_to_confirm'
]]
```

```python
df = df[
    ~(df['delta_start_to_step1'] < 0) & 
    ~(df['delta_step1_to_step2'] < 0) & 
    ~(df['delta_step2_to_step3'] < 0) & 
    ~(df['delta_step3_to_confirm'] < 0)
]
```

```python
df.to_csv('cleansed_timestamps.csv', index=False)
```
